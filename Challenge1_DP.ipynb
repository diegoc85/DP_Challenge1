```python
# Dicionário com os dados dos exames dos pacientes realizados no laboratório
exames = [
    {'nome':'Ana','sobrenome':'Souza','data':'2025-05-01','medicao':2.5,'ref_min':2.0,'ref_max':3.0},
    {'nome':'Bruno','sobrenome':'Pereira','data':'2025-05-02','medicao':1.8,'ref_min':2.0,'ref_max':3.5},
    {'nome':'Carlos','sobrenome':'Costa','data':'2025-05-03','medicao':3.2,'ref_min':2.5,'ref_max':3.0},
    {'nome':'Carlos','sobrenome':'Costa','data':'2025-05-23','medicao':2.8,'ref_min':2.5,'ref_max':3.0},
    {'nome':'Ana','sobrenome':'Souza','data':'2025-07-11','medicao':3.5,'ref_min':2.0,'ref_max':3.0},
    {'nome':'Daniela','sobrenome':'Rocha','data':'2025-05-01','medicao':2.1,'ref_min':2.0,'ref_max':2.5},
    {'nome':'Eduardo','sobrenome':'Almeida','data':'2025-05-04','medicao':2.4,'ref_min':2.2,'ref_max':2.8},
    {'nome':'Bruno','sobrenome':'Pereira','data':'2025-05-10','medicao':2.8,'ref_min':2.0,'ref_max':3.5},
    {'nome':'Igor','sobrenome':'Dias','data':'2025-07-11','medicao':2.1,'ref_min':2.9,'ref_max':3.4},
    {'nome':'Fernanda','sobrenome':'Mendes','data':'2025-05-02','medicao':1.9,'ref_min':2.0,'ref_max':2.6},
    {'nome':'Gabriel','sobrenome':'Gomes','data':'2025-05-03','medicao':3.0,'ref_min':2.8,'ref_max':3.2},
    {'nome':'Helena','sobrenome':'Barros','data':'2025-05-04','medicao':2.7,'ref_min':2.5,'ref_max':3.0},
    {'nome':'Fernanda','sobrenome':'Mendes','data':'2025-05-12','medicao':2.9,'ref_min':2.0,'ref_max':2.6},
    {'nome':'Igor','sobrenome':'Dias','data':'2025-05-01','medicao':3.1,'ref_min':2.9,'ref_max':3.4},
    {'nome':'Juliana','sobrenome':'Silva','data':'2025-05-02','medicao':2.2,'ref_min':2.0,'ref_max':2.3},
    {'nome':'Kátia','sobrenome':'Alves','data':'2025-05-03','medicao':2.9,'ref_min':2.7,'ref_max':3.1},
    {'nome':'Lucas','sobrenome':'Fernandes','data':'2025-05-04','medicao':1.7,'ref_min':1.5,'ref_max':2.0}
]

for exame in exames:
    if exame['medicao'] >= exame['ref_min'] and exame['medicao'] <= exame['ref_max']:
        exame['resultado'] = 'normal'
    else:
        exame['resultado'] = 'fora do intervalo'

def ordenar_exames(lista, campo):
    #Ordena uma cópia da lista de exames usando insertion sort
    lista_ordenada = lista.copy()
    for i in range(1, len(lista_ordenada)):
        atual = lista_ordenada[i]
        j = i - 1
        while j >= 0 and lista_ordenada[j][campo].lower() > atual[campo].lower():
            lista_ordenada[j+1] = lista_ordenada[j]
            j -= 1
        lista_ordenada[j+1] = atual
    return lista_ordenada

def busca_exame(lista, campo, valor):
    baixo, alto = 0, len(lista) - 1
    valor_lower = valor.lower()
    while baixo <= alto:
        meio = (baixo + alto) // 2
        mid_val = lista[meio][campo].lower()
        if mid_val == valor_lower:
            return meio
        if mid_val < valor_lower:
            baixo = meio + 1
        else:
            alto = meio - 1
    return -1
# Exibe os exames de forma amigável para o usuário do programa
def format_exame(exame):
    print(f"Paciente: {exame['nome']} {exame['sobrenome']}")
    print(f"Data do Exame: {exame['data']}")
    print(f"Medição: {exame['medicao']} mm")
    print(f"Intervalo de Referência: {exame['ref_min']} - {exame['ref_max']}")
    print(f"Resultado: {exame['resultado']}")
    print("-" * 40)

def listar_exames():  # Ordena os exames por nome antes de exibi-los
    ordenados = ordenar_exames(exames, 'nome')
    for exame in ordenados:
        format_exame(exame)

def buscar_exames_por_nome(nome_paciente):
    # Ordena e localiza o índice inicial com a busca binária
    ordenados = ordenar_exames(exames, 'nome')
    indice = busca_exame(ordenados, 'nome', nome_paciente)
    if indice == -1:
        print('Paciente não encontrado')
        return
    resultados = []
    esquerda = indice
    while esquerda >= 0 and ordenados[esquerda]['nome'].lower() == nome_paciente.lower():
        resultados.append(ordenados[esquerda])
        esquerda -= 1
    direita = indice + 1
    while direita < len(ordenados) and ordenados[direita]['nome'].lower() == nome_paciente.lower():
        resultados.append(ordenados[direita])
        direita += 1
    resultados = ordenar_exames(resultados, 'nome')
    for exame in resultados:
        format_exame(exame)

while True:
    print('\nBem-Vindo ao SimplIA - Sistema De Medição Patológica por IA - Laboratório DASA\n\n')
    print('1) Listar todos os exames de pacientes')
    print('2) Buscar exames pelo nome do paciente')
    print('3) Sair\n\n')
    opcao = input('Escolha uma opção: ')
    if opcao == '1':
        listar_exames()
    elif opcao == '2':
        nome = input('Digite o nome do paciente: ')
        buscar_exames_por_nome(nome)
    elif opcao == '3':
        print("Até logo!Bom descanso!")
        break
    else:
        print('OPÇÃO INVÁLIDA. TENTE NOVAMENTE!')
```
